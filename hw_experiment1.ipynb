```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, f1_score, roc_curve, auc, confusion_matrix, classification_report
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
import seaborn as sns

# Load and preprocess original data
columns = ["age", "workclass", "fnlwgt", "education", "education-num", "marital-status",
           "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss",
           "hours-per-week", "native-country", "income"]
data = pd.read_csv(r"c:\Users\USER\Documents\NTUST\2nd Sem\Data Privacy\adult\adult.data", 
                   names=columns, na_values=" ?", skipinitialspace=True)
data.dropna(inplace=True)

# Encode categorical columns for original data
le = LabelEncoder()
categorical_cols = ["workclass", "education", "marital-status", "occupation", 
                    "relationship", "race", "sex", "native-country", "income"]
for col in categorical_cols:
    data[col] = le.fit_transform(data[col])

# Exclude 'fnlwgt' from features
feature_cols_orig = [col for col in columns if col not in ["income", "fnlwgt"]]
X_orig = data[feature_cols_orig]
y_orig = data["income"]  # 0 = <=50K, 1 = >50K
X_train_orig, X_test_orig, y_train_orig, y_test_orig = train_test_split(
    X_orig, y_orig, test_size=0.3, random_state=42
)

# Load and preprocess anonymized data
anonymized_data = pd.read_csv(r"c:\Users\USER\Documents\NTUST\2nd Sem\Data Privacy\anonymized_adult.csv")

# Function to parse range strings (e.g., "[0.2-0.3]") into midpoints
def parse_range(val):
    if isinstance(val, str) and "-" in val:
        low, high = map(float, val.strip("[]").split("-"))
        return (low + high) / 2
    return float(val)  # Convert to float if already a number

# Apply parsing to anonymized QID columns
qid_cols = ["age", "education", "marital-status", "occupation", "sex"]
for col in qid_cols:
    anonymized_data[col] = anonymized_data[col].apply(parse_range)

# Ensure income is numerically encoded consistently
anonymized_data["income"] = anonymized_data["income"].map({"<=50K": 0, ">50K": 1})

# Split features and target for anonymized data (only QIDs)
X_anon = anonymized_data[qid_cols]
y_anon = anonymized_data["income"]
X_train_anon, X_test_anon, y_train_anon, y_test_anon = train_test_split(
    X_anon, y_anon, test_size=0.3, random_state=42
)

# Function to plot ROC and Confusion Matrix
def plot_metrics(y_test, y_pred, y_prob, model_name, dataset_name):
    # ROC Curve
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    roc_auc = auc(fpr, tpr)
    plt.figure()
    plt.plot(fpr, tpr, label=f"{model_name} ({dataset_name}) (AUC = {roc_auc:.2f})")
    plt.plot([0, 1], [0, 1], "k--")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"ROC Curve - {model_name} on {dataset_name} Data")
    plt.legend(loc="best")
    plt.show()

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)
    plt.title(f"Confusion Matrix - {model_name} on {dataset_name} Data")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()
    return roc_auc, cm

# --- SVM on Original Data ---
svm_orig = SVC(kernel="rbf", probability=True, random_state=42)
svm_orig.fit(X_train_orig, y_train_orig)
y_pred_orig_svm = svm_orig.predict(X_test_orig)
y_prob_orig_svm = svm_orig.predict_proba(X_test_orig)[:, 1]
acc_orig_svm = accuracy_score(y_test_orig, y_pred_orig_svm)
f1_orig_svm = f1_score(y_test_orig, y_pred_orig_svm, average="weighted")
roc_auc_orig_svm, cm_orig_svm = plot_metrics(y_test_orig, y_pred_orig_svm, y_prob_orig_svm, "SVM", "Original")
print("SVM on Original Data (without fnlwgt):")
print(f"Accuracy: {acc_orig_svm:.4f}")
print(f"F1-Score: {f1_orig_svm:.4f}")
print(f"AUC: {roc_auc_orig_svm:.4f}")
print("Confusion Matrix:\n", cm_orig_svm)
print("Classification Report:\n", classification_report(y_test_orig, y_pred_orig_svm))

# --- SVM on Anonymized Data ---
svm_anon = SVC(kernel="rbf", probability=True, random_state=42)
svm_anon.fit(X_train_anon, y_train_anon)
y_pred_anon_svm = svm_anon.predict(X_test_anon)
y_prob_anon_svm = svm_anon.predict_proba(X_test_anon)[:, 1]
acc_anon_svm = accuracy_score(y_test_anon, y_pred_anon_svm)
f1_anon_svm = f1_score(y_test_anon, y_pred_anon_svm, average="weighted")
roc_auc_anon_svm, cm_anon_svm = plot_metrics(y_test_anon, y_pred_anon_svm, y_prob_anon_svm, "SVM", "Anonymized")
print("SVM on Anonymized Data:")
print(f"Accuracy: {acc_anon_svm:.4f}")
print(f"F1-Score: {f1_anon_svm:.4f}")
print(f"AUC: {roc_auc_anon_svm:.4f}")
print("Confusion Matrix:\n", cm_anon_svm)
print("Classification Report:\n", classification_report(y_test_anon, y_pred_anon_svm))

# --- Random Forest on Original Data ---
rf_orig = RandomForestClassifier(n_estimators=100, random_state=42)
rf_orig.fit(X_train_orig, y_train_orig)
y_pred_orig_rf = rf_orig.predict(X_test_orig)
y_prob_orig_rf = rf_orig.predict_proba(X_test_orig)[:, 1]
acc_orig_rf = accuracy_score(y_test_orig, y_pred_orig_rf)
f1_orig_rf = f1_score(y_test_orig, y_pred_orig_rf, average="weighted")
roc_auc_orig_rf, cm_orig_rf = plot_metrics(y_test_orig, y_pred_orig_rf, y_prob_orig_rf, "Random Forest", "Original")
print("Random Forest on Original Data (without fnlwgt):")
print(f"Accuracy: {acc_orig_rf:.4f}")
print(f"F1-Score: {f1_orig_rf:.4f}")
print(f"AUC: {roc_auc_orig_rf:.4f}")
print("Confusion Matrix:\n", cm_orig_rf)
print("Classification Report:\n", classification_report(y_test_orig, y_pred_orig_rf))

# --- Random Forest on Anonymized Data ---
rf_anon = RandomForestClassifier(n_estimators=100, random_state=42)
rf_anon.fit(X_train_anon, y_train_anon)
y_pred_anon_rf = rf_anon.predict(X_test_anon)
y_prob_anon_rf = rf_anon.predict_proba(X_test_anon)[:, 1]
acc_anon_rf = accuracy_score(y_test_anon, y_pred_anon_rf)
f1_anon_rf = f1_score(y_test_anon, y_pred_anon_rf, average="weighted")
roc_auc_anon_rf, cm_anon_rf = plot_metrics(y_test_anon, y_pred_anon_rf, y_prob_anon_rf, "Random Forest", "Anonymized")
print("Random Forest on Anonymized Data:")
print(f"Accuracy: {acc_anon_rf:.4f}")
print(f"F1-Score: {f1_anon_rf:.4f}")
print(f"AUC: {roc_auc_anon_rf:.4f}")
print("Confusion Matrix:\n", cm_anon_rf)
print("Classification Report:\n", classification_report(y_test_anon, y_pred_anon_rf))

# --- XGBoost on Original Data ---
xgb_orig = XGBClassifier(use_label_encoder=False, eval_metric="logloss", random_state=42)
xgb_orig.fit(X_train_orig, y_train_orig)
y_pred_orig_xgb = xgb_orig.predict(X_test_orig)
y_prob_orig_xgb = xgb_orig.predict_proba(X_test_orig)[:, 1]
acc_orig_xgb = accuracy_score(y_test_orig, y_pred_orig_xgb)
f1_orig_xgb = f1_score(y_test_orig, y_pred_orig_xgb, average="weighted")
roc_auc_orig_xgb, cm_orig_xgb = plot_metrics(y_test_orig, y_pred_orig_xgb, y_prob_orig_xgb, "XGBoost", "Original")
print("XGBoost on Original Data (without fnlwgt):")
print(f"Accuracy: {acc_orig_xgb:.4f}")
print(f"F1-Score: {f1_orig_xgb:.4f}")
print(f"AUC: {roc_auc_orig_xgb:.4f}")
print("Confusion Matrix:\n", cm_orig_xgb)
print("Classification Report:\n", classification_report(y_test_orig, y_pred_orig_xgb))

# --- XGBoost on Anonymized Data ---
xgb_anon = XGBClassifier(use_label_encoder=False, eval_metric="logloss", random_state=42)
xgb_anon.fit(X_train_anon, y_train_anon)
y_pred_anon_xgb = xgb_anon.predict(X_test_anon)
y_prob_anon_xgb = xgb_anon.predict_proba(X_test_anon)[:, 1]
acc_anon_xgb = accuracy_score(y_test_anon, y_pred_anon_xgb)
f1_anon_xgb = f1_score(y_test_anon, y_pred_anon_xgb, average="weighted")
roc_auc_anon_xgb, cm_anon_xgb = plot_metrics(y_test_anon, y_pred_anon_xgb, y_prob_anon_xgb, "XGBoost", "Anonymized")
print("XGBoost on Anonymized Data:")
print(f"Accuracy: {acc_anon_xgb:.4f}")
print(f"F1-Score: {f1_anon_xgb:.4f}")
print(f"AUC: {roc_auc_anon_xgb:.4f}")
print("Confusion Matrix:\n", cm_anon_xgb)
print("Classification Report:\n", classification_report(y_test_anon, y_pred_anon_xgb))

# --- Deep Learning Model Definition ---
def build_model(input_dim):
    model = Sequential([
        Dense(64, activation="relu", input_dim=input_dim),
        Dense(32, activation="relu"),
        Dense(1, activation="sigmoid")
    ])
    model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
    return model

# --- Deep Learning on Original Data ---
model_orig = build_model(X_train_orig.shape[1])
model_orig.fit(X_train_orig, y_train_orig, epochs=10, batch_size=32, verbose=0)
loss_orig, acc_orig_dl = model_orig.evaluate(X_test_orig, y_test_orig, verbose=0)
y_prob_orig_dl = model_orig.predict(X_test_orig, verbose=0).flatten()
y_pred_orig_dl = (y_prob_orig_dl > 0.5).astype(int)
f1_orig_dl = f1_score(y_test_orig, y_pred_orig_dl, average="weighted")
roc_auc_orig_dl, cm_orig_dl = plot_metrics(y_test_orig, y_pred_orig_dl, y_prob_orig_dl, "Deep Learning", "Original")
print("Deep Learning on Original Data (without fnlwgt):")
print(f"Accuracy: {acc_orig_dl:.4f}")
print(f"F1-Score: {f1_orig_dl:.4f}")
print(f"AUC: {roc_auc_orig_dl:.4f}")
print("Confusion Matrix:\n", cm_orig_dl)
print("Classification Report:\n", classification_report(y_test_orig, y_pred_orig_dl))

# --- Deep Learning on Anonymized Data ---
model_anon = build_model(X_train_anon.shape[1])
model_anon.fit(X_train_anon, y_train_anon, epochs=10, batch_size=32, verbose=0)
loss_anon, acc_anon_dl = model_anon.evaluate(X_test_anon, y_test_anon, verbose=0)
y_prob_anon_dl = model_anon.predict(X_test_anon, verbose=0).flatten()
y_pred_anon_dl = (y_prob_anon_dl > 0.5).astype(int)
f1_anon_dl = f1_score(y_test_anon, y_pred_anon_dl, average="weighted")
roc_auc_anon_dl, cm_anon_dl = plot_metrics(y_test_anon, y_pred_anon_dl, y_prob_anon_dl, "Deep Learning", "Anonymized")
print("Deep Learning on Anonymized Data:")
print(f"Accuracy: {acc_anon_dl:.4f}")
print(f"F1-Score: {f1_anon_dl:.4f}")
print(f"AUC: {roc_auc_anon_dl:.4f}")
print("Confusion Matrix:\n", cm_anon_dl)
print("Classification Report:\n", classification_report(y_test_anon, y_pred_anon_dl))
```
